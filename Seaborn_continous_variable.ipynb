#1 Introduction
Continuous variables, also known as quantitative variables, are variables that can take on any value within a specific range or interval. They represent measurements or numerical quantities that can be expressed on a continuous scale. Continuous variables can have both integer and decimal values.

Here are a few key points about continuous variables:

**Range and Interval**: Continuous variables have a range, which defines the minimum and maximum values the variable can take. They also have an interval, which represents the smallest meaningful difference between any two values of the variable.

**Infinite Possible Values**: Continuous variables can theoretically take on an infinite number of values within their defined range. For example, variables such as age, height, weight, temperature, and income are typically considered continuous because they can take on a wide range of values.

**Measurement Precision**: Continuous variables are often measured using instruments or techniques that provide a certain level of precision. For example, a scale that measures weight in kilograms or a thermometer that measures temperature in degrees Celsius can provide precise measurements.

**Statistical Analysis**: Continuous variables are commonly used in statistical analysis. They allow for various descriptive and inferential statistical techniques, such as mean, median, standard deviation, correlation, regression analysis, and hypothesis testing.

**Visualization**: Visualizing continuous variables is important to understand their distributions, relationships, and trends. Plots such as scatter plots, line plots, histograms, kernel density estimation (KDE) plots, and box plots are commonly used to visualize continuous variables and gain insights from the data.

#2 Example with Rreal world dataset

##2.1 Import Libraries
n this section, we import the necessary libraries: seaborn and pandas. seaborn is a powerful data visualization library built on top of matplotlib, while pandas provides data manipulation and analysis functionalities.

```python
import seaborn as sns
import pandas as pd
```

##2.2 Load Dataset

Next, the "tips" dataset is loaded using the sns.load_dataset() function. This dataset is a built-in example dataset provided by Seaborn, containing information about restaurant tips, including attributes such as the total bill, tip amount, day of the week, time of the day, and party size.

```python
# Load the "tips" dataset from Seaborn
tips_df = sns.load_dataset("tips")
```

##2.3 Scatter plot
The code proceeds to create several plots to explore and visualize the "tips" dataset. First, a scatter plot is created using the sns.scatterplot() function. This plot shows the relationship between the total bill and tip amount, allowing us to understand if there is any correlation or pattern between these variables.

The plot style is then set to "darkgrid" using the sns.set() function, which applies a dark background grid to the plots, enhancing their visual appeal.

```python
# Create a scatter plot to visualize the relationship between total bill and tip amount
sns.scatterplot(data=tips_df, x="total_bill", y="tip")
sns.set(style="darkgrid")  # Set the plot style
```

##2.4 Line Plot
A line plot is created using the sns.lineplot() function to show the trend of the tip amount over the total bill. This plot helps visualize any patterns or trends in the data and provides insights into how the tip amount varies with changes in the total bill.

```python
# Create a line plot to show the trend of tip amount over total bill
sns.lineplot(data=tips_df, x="total_bill", y="tip")
```

##2.5 Distribution plot
A distribution plot, in the form of a histogram, is generated using the sns.histplot() function. This plot provides an overview of the distribution of total bill values, giving an understanding of the spread and frequency of different bill amounts.

```python
# Create a distribution plot (histogram) to understand the distribution of total bill
sns.histplot(data=tips_df, x="total_bill")
```

##2.6 KDE plot
The code then creates a KDE (Kernel Density Estimation) plot using the sns.kdeplot() function. This plot visualizes the density estimation of the tip amount, providing insights into the likelihood of different tip amounts occurring.

```python
# Create a KDE plot to visualize the density of tip amount
sns.kdeplot(data=tips_df, x="tip")
```

##2.7 Box plot

```python
# Create a box plot to compare the tip amount across different days of the week
sns.boxplot(data=tips_df, x="day", y="tip")
```
