#Introduction
 In this colab notebook, we will explore Seaborn's capabilities for visualizing relationships and patterns in data. We will use various plot types, including scatter plots, regression plots, joint plots, pair plots, and heatmaps. These plots will help us understand correlations, identify patterns, and analyze multi-dimensional datasets.


1.   **Scatter Plots:** Scatter plots are commonly used to visualize the relationship between two continuous variables. They display individual data points as dots on a two-dimensional plane, with one variable on the x-axis and the other variable on the y-axis. Seaborn's `scatterplot()` function provides additional options for customizing the appearance of scatter plots, such as adding a regression line, highlighting subgroups with different colors, or adjusting marker size.
2.   **Regression Plots:** Seaborn's `regplot()` and `lmplot()` functions allow us to create scatter plots with a fitted regression line. These plots are useful for visualizing the overall trend or relationship between two variables and can reveal any linear or non-linear patterns in the data.
3. **Joint Plots:** Joint plots combine two different plots in one figure to show the relationship between two variables. They typically include a scatter plot or regression plot on the main axes and additional plots like histograms or density plots on the marginal axes. Joint plots are created using the jointplot() function and are helpful for visualizing the distribution of individual variables along with their relationship.
4. **Pair Plots:** Seaborn's pairplot() function allows us to create a grid of scatter plots for multiple variables in a dataset. It displays the pairwise relationships between variables, creating scatter plots for each unique combination. Pair plots are particularly useful for exploring multi-dimensional datasets and identifying patterns or correlations among variables.
5. **Heatmaps:** Heatmaps are graphical representations of data using a color-coded matrix. Seaborn's heatmap() function visualizes the correlation between variables in a dataset. It uses a color scale to indicate the strength and direction of the correlation, making it easy to identify patterns and relationships within the data.

When visualizing relationships and patterns, Seaborn offers additional functionalities to enhance the visual representation. For example, the hue parameter can be used to add a categorical variable to differentiate data points or lines by color. The style parameter allows further customization, such as different marker styles or line types based on another categorical variable.


#2 Example with real world dataset

##2.1 Import libraries
In this section, we import the necessary libraries: seaborn and pandas. seaborn is a powerful data visualization library built on top of matplotlib, while pandas provides data manipulation and analysis functionalities.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

##2.2 Load Dataset
In this example, we use the popular "iris" dataset, which contains measurements of various iris flower species. We load the dataset using sns.load_dataset("iris") to obtain the DataFrame iris_df.

```python
# Load the "iris" dataset from Seaborn
iris_df = sns.load_dataset("iris")
```

```python
iris_df.head
```

##2.3 Scatter Plot
The first plot is a scatter plot that visualizes the relationship between sepal length and sepal width using sns.scatterplot(). The hue parameter is set to "species" to differentiate the data points by color based on the iris species.

```python
# Scatter plot to visualize the relationship between sepal length and sepal width
sns.scatterplot(data=iris_df, x="sepal_length", y="sepal_width", hue="species")
plt.title("Sepal Length vs Sepal Width")
```

##2.4 Regression Plot
Next, we create a regression plot using sns.regplot() to show the trend between petal length and petal width.

```python
# Regression plot to show the trend between petal length and petal width
sns.regplot(data=iris_df, x="petal_length", y="petal_width")
plt.title("Petal Length vs Petal Width")
```

##2.5 Joint Plot
We then use a joint plot with a hexagonal binning representation (kind="hex") to visualize the distribution and relationship between petal length and petal width using sns.jointplot().


```python
# Joint plot to visualize the distribution and relationship between petal length and petal width
sns.jointplot(data=iris_df, x="petal_length", y="petal_width", kind="hex")
plt.suptitle("Petal Length vs Petal Width", y=1.02)
```

##2.6 Pair Plot
The pair plot is created using sns.pairplot() to display the pairwise relationships between variables in the iris dataset. The hue parameter is set to "species" to differentiate the plots by color based on the iris species.

```python
# Pair plot to display the pairwise relationships between variables
sns.pairplot(data=iris_df, hue="species")
plt.suptitle("Pairwise Relationships in Iris Dataset", y=1.02)
```

##2.7 Heatmap
Finally, we generate a heatmap using sns.heatmap() to visualize the correlation matrix of the iris dataset. The correlation matrix is computed using iris_df.corr(), and the heatmap is annotated with correlation values and color-coded using the "coolwarm" colormap.

```python
# Heatmap to visualize the correlation matrix of the iris dataset
correlation_matrix = iris_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of Iris Dataset")
```
