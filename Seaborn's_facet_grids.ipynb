#Introduction
In this colab notebook, we will explore Seaborn's capabilities for plotting with facets and grids. Facet grids allow us to create multiple plots based on subsets or categories within the data, providing a powerful way to analyze and visualize patterns across different groups or dimensions.



1.  **Facet Grids:** Seaborn's FacetGrid class allows us to create a grid of subplots, where each subplot represents a different subset or category within the data. We can specify the variables that define the subsets or categories, and Seaborn automatically creates the grid with the appropriate number of rows and columns.
2.   **Scatter Plots with Facet Grids:** We can create scatter plots using Seaborn's scatterplot() function within a facet grid. This allows us to visualize the relationship between two continuous variables across different subsets or categories. The scatter plots in the facet grid can be further customized, such as by adding regression lines, markers, or color-coded groups.
3. **Bar Plots with Facet Grids:** Seaborn's barplot() function can be used within a facet grid to create bar plots for categorical variables. This allows us to compare values across different subsets or categories, providing insights into distributions or comparisons.

4. **Box Plots with Facet Grids:** Seaborn's boxplot() function is another powerful tool for visualizing distributions within a facet grid. Box plots provide information about the quartiles, median, and potential outliers, allowing us to compare distributions across different subsets or categories.

5. **Customizing Facet Grids:** Seaborn provides several customization options to enhance the appearance and information displayed in facet grids. We can control the layout of the grids, including the number of rows and columns, aspect ratios, and spacing. Additionally, we can modify the axes, labels, and annotations within the facet grids to provide comprehensive insights into the data.

By utilizing Seaborn's facet grids, we can effectively explore and visualize relationships, patterns, and distributions across different subsets or categories within our data. This approach enables a comprehensive analysis and facilitates data-driven decision-making.



#2 Example code with real world Dataset

##2.1 Import Libraries
In this section, we import the necessary libraries: seaborn and pandas. seaborn is a powerful data visualization library built on top of matplotlib, while pandas provides data manipulation and analysis functionalities.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

##2.2 Load Datasset
In this example, the "tips" dataset is loaded from Seaborn, which contains information about restaurant tips. The code demonstrates different plot types for categorical data visualization.

```python
# Load the "tips" dataset from Seaborn
tips_df = sns.load_dataset("tips")
```

```python
tips_df.head
```

##2.3 facet grid for scatter plots
The first facet grid creates scatter plots using sns.scatterplot() to visualize the relationship between total bill and tip amount, grouped by the time of day (lunch or dinner). Each subplot in the facet grid represents a different time of day.

```python
# Create a facet grid for scatter plots of total bill vs. tip amount, grouped by time of day
g = sns.FacetGrid(data=tips_df, col="time")
g.map(sns.scatterplot, "total_bill", "tip")
plt.suptitle("Scatter Plot: Total Bill vs. Tip Amount by Time of Day", y=1.02)
```

##2.4facet grid creates bar plots
The second facet grid creates bar plots using sns.barplot() to show the average total bill, grouped by both the day of the week and the time of day. Each subplot represents a combination of a specific day and time of day.

```python
# Create a facet grid for bar plots of average total bill, grouped by day and time of day
g = sns.FacetGrid(data=tips_df, col="day", row="time")
g.map(sns.barplot, "day", "total_bill", estimator=sum)
plt.suptitle("Bar Plot: Average Total Bill by Day and Time of Day", y=1.02)
```

##2.5 facet grid creates box plots
The third facet grid creates box plots using sns.boxplot() to display the distribution of total bill, grouped by the day of the week and the time of day. Each subplot represents a combination of a specific day and time of day.

```python
# Create a facet grid for box plots of total bill, grouped by day and time of day
g = sns.FacetGrid(data=tips_df, col="day", row="time")
g.map(sns.boxplot, "day", "total_bill")
plt.suptitle("Box Plot: Total Bill Distribution by Day and Time of Day", y=1.02)
```
