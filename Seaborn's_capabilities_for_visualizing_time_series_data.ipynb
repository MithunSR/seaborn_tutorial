#1.Introduction
In this code example, we will explore the capabilities of Seaborn for visualizing time series data. Time series data refers to data points collected over time, such as stock prices, temperature readings, or, in our case, the number of passengers on different flights.

We will be using the Seaborn library, along with pandas and matplotlib, to load and analyze the "flights" dataset. The dataset contains information about the number of passengers on various flights over a specific period.

#2 Example with real world daraset


##2.1 Importing the necessary libraries
We import the seaborn, pandas, and matplotlib.pyplot libraries to work with the data, create visualizations, and customize the plots.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

##2.2 Load Dataset
We use the sns.load_dataset() function from Seaborn to load the "flights" dataset, which contains information about the number of passengers on different flights over time. The dataset is stored in the flights_df DataFrame.

```python
flights_df = sns.load_dataset("flights")
```

##2.3 Format Dataset
We create a new column called 'date' in the DataFrame by concatenating the 'year' and 'month' columns as strings and then converting them to datetime format using pd.to_datetime(). This allows us to work with the data as a time series.

```python
# Convert the "year" and "month" columns to datetime format
flights_df['date'] = pd.to_datetime(flights_df['year'].astype(str) + '-' + flights_df['month'].astype(str))
```

##2.4  line plot
We create a line plot using sns.lineplot() to visualize the number of passengers over time. The 'date' column is used as the x-axis, the 'passengers' column is used as the y-axis, and the plot is colored in 'steelblue'. We also set the title, x-label, y-label, and rotate the x-axis labels for better readability.

```python
# Create a line plot to visualize the number of passengers over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=flights_df, x='date', y='passengers', color='steelblue', linewidth=2)
plt.title('Number of Passengers Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Passengers', fontsize=12)
plt.xticks(rotation=45)
```

##2.5 Area plot
We create an area plot using plt.fill_between() to show the distribution of passengers over time. The 'date' column is used as the x-axis, the 'passengers' column is used to fill the area, and the plot is filled with the color 'skyblue'. Again, we set the title, x-label, y-label, and rotate the x-axis labels.

```python
# Create an area plot to show the distribution of passengers over time
plt.figure(figsize=(10, 6))
plt.fill_between(flights_df['date'], flights_df['passengers'], color='skyblue')
plt.title('Distribution of Passengers Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Passengers', fontsize=12)
plt.xticks(rotation=45)
```

##2.5 Heatmap
We group the data by year and month, calculate the sum of passengers for each month, and reshape the DataFrame using groupby().sum().unstack() to create a monthly frequency view of the data. We then create a heatmap using sns.heatmap() to visualize the number of passengers, with the color scheme 'Blues'. Finally

```python
flights_monthly = flights_df.groupby([flights_df['date'].dt.year, flights_df['date'].dt.month]).sum().unstack()

plt.figure(figsize=(10, 6))
sns.heatmap(data=flights_monthly, cmap='Blues')
plt.title('Number of Passengers Heatmap', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Year', fontsize=12)
```
